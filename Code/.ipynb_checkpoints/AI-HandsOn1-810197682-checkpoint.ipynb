<h2 style = "font-size: 25px;">Reading Test File Content</h2>

```python
def readInput(testFile) :
    file = open(testFile, 'r+')
    fileList = file.readlines()
    fileList = [s.replace('\n', '') for s in fileList]
    
    [days, doctors] = [int(i) for i in fileList[0].split()]
    maxCapacity = int(fileList[1])
    
    allShifts = []
    for i in range(2, days + 2):
        dayRequirements = fileList[i].split()
        morningReqs = [int(i) for i in dayRequirements[0].split(",")]
        eveningReqs = [int(i) for i in dayRequirements[1].split(",")]
        nightReqs = [int(i) for i in dayRequirements[2].split(",")]
        
        allShifts.append((morningReqs, eveningReqs, nightReqs))

    file.close()
    return [days, list(range(doctors)), maxCapacity, allShifts]
```

<h2 style = "font-size: 25px;">Job Scheduler</h2>

```python
import time
import random
import numpy as np

class JobScheduler:
    def __init__(self, fileInfo):
        self.days = fileInfo[0]
        self.doctors = len(fileInfo[1])
        self.doctorsIds = fileInfo[1]
        self.maxCapacity = fileInfo[2]
        self.allShifts = fileInfo[3]

        self.popSize = 300
        self.crossOverPoints = self.days
        self.elitismPercentage = 16
        self.pc = 0.8
        self.pm = 0.001
        self.chromosomeLength = self.days * self.doctors * 3
        self.chromosomes = self.generateInitialPopulation()
        self.sampleProb = np.array(list(reversed(range(self.popSize)))) / (self.popSize * (self.popSize - 1) / 2)  #Formula has taken from the Internet
        
    def generateInitialPopulation(self):
        population = [] 
        
        for i in range(self.popSize):
            chromosome = [int(random.randint(0, 1)) for i in range(self.chromosomeLength)]
            population.append((chromosome, self.calculateFitness(chromosome)))
        
        return population
        
    def onePointCross(self, point, chromosome1, chromosome2):
        new_chromosome1 = chromosome1[:point] + chromosome2[point:]
        new_chromosome2 = chromosome2[:point] + chromosome1[point:]
        return new_chromosome1, new_chromosome2
    
    def crossOver(self, chromosome1, chromosome2):
        if random.uniform(0, 1) >= self.pc:
            return [chromosome1, chromosome2]

        points = np.random.choice(range(self.chromosomeLength), size = self.crossOverPoints, replace = False)

        for p in points:
            new_chromosome1, new_chromosome2 = self.onePointCross(p, chromosome1[0], chromosome2[0])
            chromosome1 = (new_chromosome1, -1)
            chromosome2 = (new_chromosome2, -1)
        
        return [chromosome1, chromosome2]

    def mutate(self, chromosome):
        newChromosome = []
        
        for i in range(self.chromosomeLength):
            newChromosome.append(abs(chromosome[0][i] - (random.uniform(0, 1) < self.pm))) 
    
        return (newChromosome, -1)

    def reshapeChromosome(self, chromosome):
        daysData = [chromosome[(3*i*self.doctors) : (3*(i + 1)*self.doctors)] for i in range(self.days)]
        shiftsData = []
        for d in daysData:
            shiftsData.append([d[j*self.doctors : (j + 1)*self.doctors] for j in range(3)])
        return shiftsData

    def calculateFitness(self, chromosome):
        totalCost = 0
        shiftsData = self.reshapeChromosome(chromosome)

        for day in range(self.days):
            for shift in range(3):
                doctorsCount = np.sum(shiftsData[day][shift])
                if (doctorsCount < self.allShifts[day][shift][0] ) or (self.allShifts[day][shift][1] < doctorsCount):
                    totalCost += abs(self.allShifts[day][shift][0] - doctorsCount)
            if day > 0:
                for doctor in range(self.doctors):
                    if shiftsData[day - 1][2][doctor] == 1 and (shiftsData[day][0][doctor] == 1 or shiftsData[day][1][doctor] == 1):
                        totalCost += 1
                    if day < self.days - 1 and shiftsData[day - 1][2][doctor] == 1 and shiftsData[day][2][doctor] == 1 and shiftsData[day + 1][2][doctor] == 1:
                        totalCost += 1

        for doctor in range(self.doctors):
            workDays = 0
            for day in range(self.days):
                for shift in range(3):
                    if shiftsData[day][shift][doctor] == 1:
                        workDays += 1
            if workDays > self.maxCapacity:
                totalCost += (workDays - self.maxCapacity)
        
        return totalCost

    def generateNewPopulation(self):
        self.chromosomes.sort(key = lambda x : x[1])
        
        if(self.chromosomes[0][1] == 0):
            return True

        matingPool = self.chromosomes[:int(self.popSize * self.elitismPercentage / 100)]
        
        while(len(matingPool) < self.popSize):    
            parentIndex_1, parentIndex_2 = np.random.choice(len(self.chromosomes), 2, p = self.sampleProb)
            children1, children2 = self.crossOver(self.chromosomes[parentIndex_1], self.chromosomes[parentIndex_2])
            children1 = self.mutate(children1)
            children2 = self.mutate(children2)
            matingPool.append((children1[0], self.calculateFitness(children1[0])))
            matingPool.append((children2[0], self.calculateFitness(children2[0])))

        self.chromosomes = matingPool
        return False
    
    def schedule(self): 
        goal_reached = False
        while (goal_reached == False):
            goal_reached = self.generateNewPopulation()
        return 
    
    def print_schedule(self):
        schedule = ""
        shiftsData = self.reshapeChromosome(self.chromosomes[0][0])

        for day in range(self.days):
            result = ""
            for shift in range(3):
                first = True
                for doctor in range(self.doctors):
                    if shiftsData[day][shift][doctor] == 1:
                        if first == False:
                            result += ","
                        first = False
                        result += str(self.doctorsIds[doctor])
                if first == True:
                    result += "empty"
                if shift < 2:
                    result += " "
            schedule += result + "\n"
            
        return schedule
```

<h2 style = "font-size: 25px;">Execution</h2>

```python
fileInfo1 = readInput(testFile1)

for i in range(2):
    start = time.time()
    scheduler = JobScheduler(fileInfo1)
    scheduler.schedule()
    end = time.time()

    print("test 1: ", '%.2f'%(end - start), 'sec')

print()

answer = scheduler.print_schedule()
print(answer)
outfile1 = open("output1.txt", "w")
outfile1.write(answer)
outfile1.close()
```
